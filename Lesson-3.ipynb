```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
literacy = pd.read_csv("UNdata_Export_20190226_203636225.csv")
print(literacy.head(n=4))
```

```python
sns.relplot(x="Year", y="Value", col= "Subgroup",  data= literacy)
```

```python
sns.lmplot(x="Year", y="Value", col= "Subgroup", data= literacy)
```

```python
sns.lmplot (x="Year", y="Value",col="Subgroup", data=literacy, x_estimator=np.mean)
```

```python
sns.catplot(x="Year", y= "Value", col= "Subgroup",kind="box", data= literacy)
```
