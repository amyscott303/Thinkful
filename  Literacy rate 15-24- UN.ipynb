```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics 
import seaborn as sns

```

```python
#This dataset is from the UN Data that the government has collected, for the literacy of females and males ages 15-24 
#years old. This data intrigued me because of all the different countries policies for education of the genders.
# In order to get clear representation of the data and display it properly it needs to be broken down by gender and 
#year so that we can compare the differences not only between genders by year but also to see if literacy rates improve
# or decline as the years progress. 
```

```python
#Within this data I wanted to see if there were large discrepencies between the two genders, knowing that in many
#countries females are still fighting to recieve equal education. From here I also wanted to see if there were changes
#year over year for both genders or did it remain the same?
```

```python
literacy = pd.read_csv("UNdata_Export_20190226_203636225.csv")
print(literacy.head(n=4))
```

```python
grouped_by = literacy.groupby(['Subgroup','Year'])
row_count = grouped_by.Value.count()
msk = (row_count >= 10)

```

```python
#The table below displays the data broken down by gender and then year. Within each year it displays the mean, standard
#deviation, and the break down of the data. I hid the data that had a data count under 10 to pull out the outliers. 
```

```python
grouped_by.Value.describe()[msk]
```

```python
sns.relplot(x="Year", y="Value", col= "Subgroup",  data= literacy)
```

```python
#These charts show the percentages of literacy rate per year broken down by gender. As you can see for both genders 
#most of the literacy rates are above 80%. The early 90s had more females rating below the 50% literacy rate,as the 
# years go on there are less countries below the 20% mark. It looks like 2000 was a good year for literacy for females,
#a large jump in improved literacy rates. It looks like there are 2 countries that are still between 20-40%. 
#For males there aren't any data points below 20%. You'll notice that there data points are less dispersed and mostly
#range between 80-100%.
```

```python
sns.lmplot(x="Year", y="Value", col= "Subgroup", data= literacy)
```

```python
sns.lmplot (x="Year", y="Value",col="Subgroup", data=literacy, x_estimator=np.mean)
```

```python
#This graph shows the mean percentage point for both genders. The charts show similarity in the means, telling us that
#those lower percentage points are most likely outliers in the data. Interestingly this shows that for both genders 1998
# had lower means by quite a bit then previous surveys.
```

```python
sns.catplot(x="Year", y= "Value", col= "Subgroup",kind="box", data= literacy)
```

```python
#I like this chart as it shows the means, where the majority of the data lies and the true outliers in the data. 
#For the females there are more outliers which shows that there still are a few countries within the UN that have poor
#education and literacy programs for females. The boxes for females are also spread out more displaying that their data points 
#for 25%-75% are more spread out in comparison to their male counterparts. 
```

```python
# This data shows that for the most part the countries in the UN are on the same page for genders with the median 
#percentages being close across the years. The data does show that there are more countries with lower literacy rates
#for females, however the data also shows there is progress being made in the literacy rates through the years for 
#both genders. 
```
