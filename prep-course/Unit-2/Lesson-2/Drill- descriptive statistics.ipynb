```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
binomial = np.random.binomial (100,0.5,1000)
plt.hist(binomial)
plt.show()
```

```python
plt.hist(binomial)
plt.axvline(binomial.mean(), color = 'b', linestyle = 'solid', linewidth= 2)
plt.axvline(binomial.mean()+binomial.std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline (binomial.mean()- binomial.std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.show()
```

```python
#the std to the right of the mean is higher and does not create a normal curve of the data. 
```

```python
import numpy as np
import matplotlib.pyplot as plt
rand1 = np.random.normal(5,.5, 100)
rand2 = np.random.normal(10,1, 100)
rand3 = rand1+rand2

mean = np.mean(rand3)
sd = np.std(rand3)

plt.hist(rand3)
plt.axvline(x=mean,color='black')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show()
```
